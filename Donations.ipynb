```python
#Importing Packages
import csv
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn import svm
from sklearn.ensemble import RandomForestRegressor
```

```python
#Reading Data set File
data=pd.read_csv("C:/Users/gadir/Desktop/Donations.csv")
```

```python
#Viewing the dataset file
data
```

```python
#Performing One-hot encoding to convert categorical variables to Numerical Variables
data1=pd.concat([data,pd.get_dummies(data['Job Titles'], prefix='JobTitles'),pd.get_dummies(data['Department'], prefix='Department')],axis=1)
```

```python
#Viewing the dataset after One-hot encoding is performed
data1
```

```python
#Replacing String characters with integers
data1['Salary or Hourly'] = data1['Salary or Hourly'].replace('Salary','1') 
data1['Salary or Hourly'] = data1['Salary or Hourly'].replace('Hourly','1') 
```

```python
#Replacing String characters with integers
data1['Full or Part-Time'] = data1['Full or Part-Time'].replace('F','1') 
data1['Full or Part-Time'] = data1['Full or Part-Time'].replace('P','0') 
```

```python
#Converting data types of object columns to integer type
data1['Salary or Hourly'] = data1['Salary or Hourly'].astype(int)
data1['Full or Part-Time'] = data1['Full or Part-Time'].astype(int)
data1['Donations'] = data1['Donations'].astype(int)
```

```python
#Making Donations Column as Target Variable 
y=data1['Donations']
#Removing Unwanted columns
X=data1.drop(['Job Titles', 'Department','Donations','Name'], axis=1)
```

```python
#Splitting the entire data set into train data and test data in 70:30 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Performing XGBOOST Model and Evaluating the performance with respect to Root Mean Squared Error
xg_reg = xgb.XGBRegressor()
xg_reg.fit(X_train, y_train)
preds = xg_reg.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```

```python
#Performing SVM Model and Evaluating the performance with respect to Root Mean Squared Error
svm_reg = svm.SVR()
svm_reg.fit(X_train, y_train)
predssvm = svm_reg.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, predssvm))
print("RMSE: %f" % (rmse))
```

```python
#Performing Random Forest  Model and Evaluating the performance with respect to Root Mean Squared Error
regr = RandomForestRegressor()
regrr=regr.fit(X_train, y_train)
predsrr = regrr.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, predsrr))
print("RMSE: %f" % (rmse))
```
